```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
r=requests.get('https://en.wikipedia.org/wiki/S%26P_100')
soup=bs(r.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
title=soup.find('title')
print(title)
```

```python
tables=soup.find_all('table')
print(len(tables))
print()
print(tables[0].prettify())
```

```python
sortable_tables=soup.find_all('table',{'class':'sortable'})
print(len(sortable_tables))
print(sortable_tables[0].prettify())
```

```python
sortable_tables[0].attrs
```

```python
text=sortable_tables[0]
print(text.text)
```

```python
sortable_table_classes=[table.attrs['class'] for table in sortable_tables]
for class_list in sortable_table_classes:
    print(class_list)
```
